```python
# Importation des librairies qu'on aura à utiliser
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff
```

```python
# chargement du jeu de données et affichage de la structure
ligne, data = loadarff("C:/Program Files/Weka-3-8-6/data/supermarket.arff")
DFrame = pd.DataFrame(ligne)
long = len(DFrame)
DFrame.head()
print("la longueur du jeu de donnee est de: ", long)
print()
print("LLe nombre de lignes et de colonne: ", DFrame.shape)
DFrame
```

```python
# La liste des colonnes du supermarket
print(DFrame.columns)
```

```python
# Ajoutons ces données dans une liste pour enfin les transformer avec transactionEncoder
list = []
for a in range(0, long):
    list.append([str(DFrame.values[a, b]) for b in range(0, 217)])
list=np.array(list)
print(list.shape)
```

Trnasformons nos donnees en un tableau croissant binair car la librairie que nous
allons utiliser prend en entree un tableau binaire

```python
# transformons nos donnees en un tableau binaire car la librairie que nous allons utiliser
#ne prend en parametre que true ou false ou 1 et 0 
encod = TransactionEncoder()
trans = encod.fit(list).transform(list)
dat = pd.DataFrame(trans, columns=encod.columns_)
#dat
print("Affichons les 15 premieres ligne et les 2 premieres lignes de notre tableau binaire")  
dat.iloc[:15,:2]
```

```python
#Affichons les itemsets frequents avec mun_sup = 0.03,

itemset_freq = fpgrowth(dat, min_support=0.03, use_colnames=True)
#print(itemset_freq.columns)
print(itemset_freq)
```


Le dataframe itemset_freq contient deux colonnes la première nommée Support et la deuxième nommée itemsets

```python
print('regardons la structure c est-à-dire le nombre de ligne et de colonne de ce dataframe')
print(itemset_freq.shape)
```

```python
#trouvons les regles d'association avec les itemsets frequent via l'algo FPGrowth
Regles_Association_fpg =association_rules(itemset_freq, metric='confidence', min_threshold=0.4)
print(Regles_Association_fpg.shape)
```
