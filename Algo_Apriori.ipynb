```python
import numpy as np
import pandas as pd
from scipy.io.arff import loadarff
from apyori import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
# Ici on charge nos donnees, on les met dans deux variables (ligne et donnees), On cree une dataframe qui recois nos donnees puis on
#regarde la longeur avec la fonction lenth, on affiche l'entete des donnees avec la fonction head  enfin on regarde la structure du jeu de donnees
ligne, donnees = loadarff("C:/Program Files/Weka-3-8-6/data/supermarket.arff")
DFrame = pd.DataFrame(ligne)
long = len(DFrame)
DFrame.head()
print(long)
DFrame.shape
DFrame
```

```python
# Avant de transforrmer ces données en tableau binair pour pouvoir faire l'extraction, voyons les noms des colonnes au préalable
print(DFrame.columns)
```

```python
# Ajoutons ces données dans une liste pour enfin les transformer avec transactionEncoder
list = []
for a in range(0, long):
    list.append([str(DFrame.values[a, b]) for b in range(0, 217)])
list=np.array(list)
print(list.shape)
```

```python
#transformons nos donnees en un tableau binaire pour pouvoir les utiliser avec la librairie qui prend en entrée bool ou bin 
encod = TransactionEncoder()
trans = encod.fit(list).transform(list)
dat = pd.DataFrame(trans, columns=encod.columns_)
dat
#dat.shape
```

```python
# Utilisons la fonction apriori pour trouver les itemsets de supermarket

itemset_freq = apriori(dat,min_support=0.15, max_len=4,use_colnames=True)
#itemset_freq = apriori(dat, min_support=0.030, use_colnames=True)
print(itemset_freq)
```

Le resultat nous que notre dataframe est vide, je ne crois pas à cela, dans la ligne qui suivrent nous allons utiliser un min_sup = 0.03

```python
# Maintenant essayons de trouver les regles d'association avec les itemsets frequents trouvés via l'algo apriori 
regles_assoc = association_rules(itemset_freq, metric="confidence", min_threshold=0.4)
regles_assoc
```
